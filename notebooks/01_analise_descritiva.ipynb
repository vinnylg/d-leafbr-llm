## Analise Descritiva da tabela "DADOS_ABERTOS_MEDICAMENTOS.csv"

```python
import leafbr.datasets.utils as u
all_drugs = u.get_all_drugs()
all_drugs
```

### Quantos medicamentos estão na tabela e qual a situação desses medicamentos?

```python
c = all_drugs['situacao_registro'].value_counts().to_frame()
c['%'] = (c/c.sum() * 100)
c.loc['TOTAL'] = c.sum(0)
c['count'] = c['count'].astype(int)
c
```

### Dos medicamentos validos ou ativos, qual a distribuição das categorias regulatorias?

```python
validos = u.get_valid_drugs()
c = validos['categoria_regulatoria'].value_counts().to_frame()
c['%'] = (c/c.sum() * 100)
c
```

### Dos medicamentos validos, quantos possuem classe terapeutica compostas? 

```python
c = validos['classe_terapeutica_composto'].value_counts().to_frame()
c['%'] = (c/c.sum() * 100)
c
```

### Qual a distribuição da classes terapeutica não composta?

```python
c = validos.loc[~validos['classe_terapeutica_composto'],'classe_terapeutica'].value_counts().to_frame()
c['%'] = (c/c.sum() * 100)
c#.head(5)
```

```python
len(validos.loc[validos['classe_terapeutica_composto'],'classe_terapeutica'].unique())
```

### Dos medicamentos validos, quantos possuem principio ativo composto?

```python
c = validos['principio_ativo_composto'].value_counts().to_frame()
c['%'] = (c/c.sum() * 100)
c
```

### Qual a distribuição dos principios ativos não composto?

```python
c = validos.loc[~validos['principio_ativo_composto'],'principio_ativo'].value_counts().to_frame()
c['%'] = (c/c.sum() * 100)
c#.head(5)
```

```python
len(validos.loc[validos['principio_ativo_composto'],'principio_ativo'].unique())
```

### Dos medicamentos válidos, quantos foram baixados e não baixados?

```python
validos_baixados = u.get_download_leaflets()

validos_baixados['downloaded'] = False
validos_baixados.loc[validos_baixados['filepath'].notna(),'downloaded'] = True

c = validos_baixados['downloaded'].value_counts().to_frame()
c['%'] = (c/c.sum() * 100)
c
```

### Dos medicamentos não baixados, quantos não possui nenhum correspondente de outra categoria? Quantos são compostos?

```python
downloaded = validos_baixados.loc[validos_baixados['downloaded']].copy()
not_downloaded = validos_baixados.loc[~validos_baixados['downloaded']].copy()

not_downloaded['tem_similar'] = not_downloaded['principio_ativo'].isin(downloaded['principio_ativo'])

for i, df in not_downloaded.loc[~not_downloaded['tem_similar']].iterrows():
    if '+' in df['principio_ativo']:
        pass
    else:
        baixado = downloaded.loc[downloaded['principio_ativo'].str.contains(df['principio_ativo'], regex=False)]
        if len(baixado) > 0:
            not_downloaded.loc[i,'tem_similar'] = True


"%d não possuem nenhum principio ativo correspondente baixado, sendo que %d possuem principio ativo composto" % (
    len(not_downloaded.loc[~not_downloaded["tem_similar"]]),
    len(
        not_downloaded.loc[
            ~not_downloaded["tem_similar"]
            & not_downloaded["principio_ativo"].str.contains("+", regex=False)
        ]
    ),
)
```

### Quantos PDFs não retornaram nada?

```python
pre_base = u.get_prebase()
pre_base.loc[pre_base['texto_bula'].isna()]
```

### Quantos arquivos PDF não estão condizentes com o nome ou principio ativo da bula na tabela?

```python
c = pre_base['bula_valida'].value_counts().to_frame()
c['%'] = (c/c.sum() * 100)
c
```

### Quantos desses são duplicados?

```python
pre_base['bula_duplicada'] = pre_base.duplicated('hash',keep=False)
c = pre_base['bula_duplicada'].value_counts().to_frame()
c['%'] = (c/c.sum() * 100)
c
```

### Qual a relação entre os duplicados e invalidos?

```python
pre_base.groupby(['bula_duplicada','bula_valida']).size().unstack().fillna(0).astype(int)
```

### Das bulas invalidas ou duplicadas, quantos não possuem nenhum correspondente principio ativo correspondente valido? Quantos são compostos?

```python
validas = pre_base.loc[pre_base["bula_valida"] & ~pre_base["bula_duplicada"]].copy()

# invalidas = pre_base.loc[~pre_base["bula_valida"]].copy()
invalidas = pre_base.loc[~(pre_base["bula_valida"] & ~pre_base["bula_duplicada"])].copy()

invalidas['tem_similar'] = invalidas['principio_ativo'].isin(validas['principio_ativo'])

for i, df in invalidas.loc[~invalidas['tem_similar']].iterrows():
    if '+' in df['principio_ativo']:
        pass
    else:
        baixado = validas.loc[validas['principio_ativo'].str.contains(df['principio_ativo'], regex=False)]
        if len(baixado) > 0:
            invalidas.loc[i,'tem_similar'] = True


"Dos %d arquivos invalidos, %d não possuem nenhum principio ativo correspondente valido, sendo que %d possuem principio ativo composto" % (
    len(invalidas),
    len(invalidas.loc[~invalidas["tem_similar"]]),
    len(
        invalidas.loc[
            ~invalidas["tem_similar"]
            & invalidas["principio_ativo"].str.contains("+", regex=False)
        ]
    ),
)
```
